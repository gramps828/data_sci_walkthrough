```python
import pandas as pd
import numpy as np
import itertools
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('news.csv')
```

```python
labels = df.label
#titles = df.title
#titles.head()
labels.head()
```

```python
#Training and testing data sectioning
x_train, x_test, y_train, y_test = train_test_split(df['text'], labels, test_size = 0.2, random_state = 7)
```

```python
#Initalize Tfidf
tfidf_vector = TfidfVectorizer(stop_words='english', max_df = 0.7)

#Fitting Data
tfidf_train = tfidf_vector.fit_transform(x_train)
tfidf_test = tfidf_vector.transform(x_test)
```

```python
#Init Passive Agressive Classifier
pac = PassiveAggressiveClassifier(max_iter = 50)
pac.fit(tfidf_train, y_train)

#Test set prediciton
y_pred = pac.predict(tfidf_test)
score = accuracy_score(y_test, y_pred)
print(f'Accuracy: {round(score*100,2)}%')
```

```python
#Build Confusion Matrix
confusion_matrix(y_test, y_pred, labels = ['FAKE', 'REAL'])
```
